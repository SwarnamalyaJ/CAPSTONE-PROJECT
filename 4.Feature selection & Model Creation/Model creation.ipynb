```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1=pd.read_csv("Preprocess_Bank_churners.csv")
df1
```

```python
df1.columns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Assuming the dataset is stored in a DataFrame named 'df'

# Split the dataset into independent (X) and dependent (y) variables
independent = df1[['Customer_Age', 'Gender', 'Dependent_count', 'Education_Level', 'Marital_Status',
                  'Income_Category', 'Card_Category', 'Months_on_book', 'Total_Relationship_Count',
                  'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
                  'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Trans_Ct',
                  'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']]

dependent = df1['Attrition_Flag']

# Apply StandardScaler to the independent variables
scaler = StandardScaler()
independent_scaled = scaler.fit_transform(independent)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(independent_scaled, dependent, test_size=0.33, random_state=42)

# Verify the shapes of the resulting splits
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# Ensure 'Attrition_Flag' is treated as a categorical (binary) variable
y_train = y_train.astype('int')

# Check if the conversion was successful
print(y_train.unique())

# Now proceed with training the Logistic Regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Create and train the model
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train, y_train)

# Make predictions
y_pred_logreg = logreg.predict(X_train)

# Evaluate the model
print("Logistic Regression Accuracy:", accuracy_score(y_train, y_pred_logreg))
print(classification_report(y_train, y_pred_logreg))

```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Create and train the Naive Bayes model
nb = GaussianNB()
nb.fit(X_train, y_train)

# Make predictions
y_pred_nb = nb.predict(X_train)

# Evaluate the model
print("Naive Bayes Accuracy:", accuracy_score(y_train, y_pred_nb))
print(classification_report(y_train, y_pred_nb))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create and train the KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Make predictions
y_pred_knn = knn.predict(X_train)

# Evaluate the model
print("KNN Accuracy:", accuracy_score(y_train, y_pred_knn))
print(classification_report(y_train, y_pred_knn))
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Create and train the SVC model
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)

# Make predictions
y_pred_svc = svc.predict(X_train)

# Evaluate the model
print("SVC Accuracy:", accuracy_score(y_train, y_pred_svc))
print(classification_report(y_train, y_pred_svc))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create and train the Decision Tree model
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

# Make predictions
y_pred_dt = dt.predict(X_train)

# Evaluate the model
print("Decision Tree Accuracy:", accuracy_score(y_train, y_pred_dt))
print(classification_report(y_train, y_pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Create and train the Random Forest model
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf.predict(X_train)

# Evaluate the model
print("Random Forest Accuracy:", accuracy_score(y_train, y_pred_rf))
print(classification_report(y_train, y_pred_rf))
```

## Conclusion

## Both the Random Forest and Decision Tree models achieved 100% accuracy, indicating their strong performance on the given dataset. This suggests that both models effectively capture the underlying patterns and trends in the data.
